## Roberta Pretrained Model

```python
from transformers import AutoTokenizer
from transformers import AutoModelForSequenceClassification
from scipy.special import softmax
```

```python
MODEL = f"cardiffnlp/twitter-roberta-base-sentiment"
tokenizer = AutoTokenizer.from_pretrained(MODEL)
model = AutoModelForSequenceClassification.from_pretrained(MODEL)
```

```python
print(ex)
```

```python
sia.polarity_scores(ex)
```

```python
encoded_text = tokenizer(ex, return_tensors='pt')
encoded_text
```

```python
output_model = model(**encoded_text)
output_model
```

```python
roberta_scores = output_model[0][0].detach().numpy()
roberta_scores = softmax(roberta_scores)
roberta_scores
```

```python
scores_dict = {
    'roberta_neg':roberta_scores[0],
    'roberta_neu':roberta_scores[1],
    'roberta_pos':roberta_scores[2]
}
print(scores_dict)
```

```python
def polarity_scores_roberta(text):
    encoded_text = tokenizer(text, return_tensors='pt')
    output_model = model(**encoded_text)
    roberta_scores = output_model[0][0].detach().numpy()
    roberta_scores = softmax(roberta_scores)
    scores_dict = {
        'roberta_neg':roberta_scores[0],
        'roberta_neu':roberta_scores[1],
        'roberta_pos':roberta_scores[2]
    }
    return scores_dict
```

```python
try:
    test = polarity_scores_roberta(df['Text'][82])
    print(df['Text'][82])
    print(test)
except (RuntimeError,IndexError):
    print("Error")
try:
    test = polarity_scores_roberta(df['Text'][117])
    print(df['Text'][117])
    print(test)
except (RuntimeError,IndexError):
    print("Error")
try:
    test = polarity_scores_roberta(df['Text'][267])
    print(df['Text'][267])
    print(test)
except (RuntimeError,IndexError):
    print("Error")
```

```python
res = {}
count = 0
for i, row in tqdm(df.iterrows(), total=len(df)):
    try:
        text = row['Text']
        myid = row['Id']
        vader_result = sia.polarity_scores(text)
        vader_result_rename = {}
        for key, value in vader_result.items():
            vader_result_rename[f"vader_{key}"] = value
        roberta_result = polarity_scores_roberta(text)
        both = {**vader_result_rename, **roberta_result}
        res[myid] = both
    except (RuntimeError,IndexError):
        count += 1

print(count)
```

```python
results_df = pd.DataFrame(res).T
results_df = results_df.reset_index().rename(columns={'index':'Id'})
results_df = results_df.merge(df, how='left')
```
